```python
import pandas as pd
movies_data = pd.read_csv('https://raw.githubusercontent.com/podkovko/ds_course/refs/heads/main/Data%20Science%20Course/PY_12%20%D0%9F%D1%80%D0%BE%D0%B4%D0%B2%D0%B8%D0%BD%D1%83%D1%82%D1%8B%D0%B5%20%D0%BC%D0%B5%D1%82%D0%BE%D0%B4%D1%8B%20%D1%80%D0%B0%D0%B1%D0%BE%D1%82%D1%8B%20%D1%81%20%D0%B4%D0%B0%D0%BD%D0%BD%D1%8B%D0%BC%D0%B8%20%D0%B2%20Pandas/data/ratings_movies.csv')
movies = movies_data.copy()
import re 
def get_year_release(arg):
    candidates = re.findall(r'\(\d{4}\)', arg) 
    if len(candidates) > 0:
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        return None

# Задание 8.1
movies['year release'] = movies['title'].apply(get_year_release)
movies['year release'].isna().sum()

# Задание 8.2
mean_ratings_1999 = movies[movies['year release'] == 1999].groupby(by='title')['rating'].mean().sort_values()

# Задание 8.3
genre_ratings_2010 = movies[movies['year release'] == 2010].pivot_table(
    values='rating',
    index='genres',
)
genre_ratings_2010 = genre_ratings_2010.sort_values('rating')

# Задание 8.4
user_genres = movies.pivot_table(
    index='userId',
    columns='genres',
    aggfunc='count'
)
user_genre_count = pd.DataFrame()
user_genre_count['Count'] = user_genres.notna().sum(axis=1).sort_values()

# Задание 8.5
user_analysis = movies.groupby('userId')['rating'].agg(
    ['count', 'mean']
).sort_values(by=['count', 'mean'], ascending=[True, False])

# Задание 8.6
mask1 = movies['year release'] == 2018
genre_analysis = movies[mask1].groupby('genres')['rating'].agg(
    ['mean', 'count']
).sort_values(by=['mean'])
mask2 = genre_analysis['count'] > 10
genre_analysis = genre_analysis[mask2]

# Задание 8.7
movies['date'] = pd.to_datetime(movies['date'])
movies['year_rating'] = movies['date'].dt.year
year_genre = movies.pivot_table(
    values='rating',
    index='genres',
    columns='year_rating'
)
display(year_genre)

year_genre.loc['Action|Adventure'].min()
year_genre.loc['Action|Adventure|Animation|Children|Comedy|IMAX'].sort_values(ascending=False)
year_genre.loc['Comedy']
```
